```python
import pandas as pd
import numpy as np
```

```python
mvps=pd.read_csv(r'https://github.com/shreyasdeodhare/Datasets101/raw/main/mvps.csv')
```

```python
mvps
```

```python
mvps=mvps[["Player","Year","Pts Won","Pts Max","Share"]]
```

```python
mvps
```

```python
players=pd.read_csv(r"https://github.com/shreyasdeodhare/Datasets101/raw/main/players.csv")
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players["Player"]=players["Player"].str.replace("*","",regex=False)
```

```python
players.groupby(["Player","Year"])
```

```python
def single_row(df):
  if df.shape[0]==1:
     return df
  else:
    row=df[df["Tm"]=="TOT"]
    row["Tm"]=df.iloc[-1,:]["Tm"]
    return row

players=players.groupby(["Player","Year"]).apply(single_row)
```

```python
players.index=players.index.droplevel()
```

```python
players.index=players.index.droplevel()
```

```python
players
```

```python
combine=players.merge(mvps,how="outer",on=["Player","Year"])
```

```python
combine.head(5)
```

```python
teams=pd.read_csv(r"https://github.com/shreyasdeodhare/Datasets101/raw/main/teams.csv")
```

```python
teams=teams[~teams["W"]. str.contains("Division" )]
```

```python
teams.head(30)
```

```python
teams["Team"]=teams["Team"].str.replace("*","",regex=False)
```

```python
teams.head(5)
```

```python
nicknames={}

with open("nicknames.txt") as f:
   lines=f.readlines()
   for line in lines[1:]:
     abbrev,name=line.replace("\n","").split(",")
     nicknames[abbrev]=name
```

```python
nicknames
```

```python
combine.head()
```

```python
combine["Team"]=combine["Tm"].map(nicknames)
```

```python
stats=combine.merge(teams,how="outer",on=["Team","Year"])
```

```python
stats
```

```python
del stats["Unnamed: 0"]
```

```python
stats=stats.apply(pd.to_numeric,errors="ignore")
```

```python
stats["GB"]=stats["GB"].str.replace("—","0")
```

```python
stats["GB"].unique()
```

```python
stats["GB"]=pd.to_numeric(stats["GB"])
```
